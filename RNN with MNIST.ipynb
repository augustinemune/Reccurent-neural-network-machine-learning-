```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, LSTM, CuDNNLSTM
```

```python
mnist = tf.keras.datasets.mnist
```

```python
(x_train, y_train),(x_test, y_test) = mnist.load_data()
x_train = x_train/255.0
x_test = x_test/255.0
```

```python
from tensorflow.keras.callbacks import TensorBoard
NAME = "mnist-RNN"
tensorboard = TensorBoard(log_dir="mnistlogs/{}".format(NAME))
```

```python
model = Sequential()
model.add(LSTM(64, input_shape=(x_train.shape[1:]), activation='relu', return_sequences=True))
model.add(Dropout(0.2))

model.add(LSTM(64, activation='relu', return_sequences=True))
model.add(Dropout(0.1))

model.add(LSTM(32, activation='relu'))
model.add(Dropout(0.1))

model.add(Dense(32, activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(10, activation='softmax'))

opt = tf.keras.optimizers.Adam(lr=1e-3, decay=1e-5)

model.compile(
    loss='sparse_categorical_crossentropy',
    optimizer=opt,
    metrics=['accuracy'],
)

model.fit(x_train, y_train, epochs=3, callbacks=[tensorboard], validation_data=(x_test, y_test))
```

## For GPU VERSION

#### Use the CuDNNLSTM instend of LSTM

```python
model = Sequential()
model.add(CuDNNLSTM(64, input_shape=(x_train.shape[1:]), return_sequences=True))
model.add(Dropout(0.2))

model.add(CuDNNLSTM(64, return_sequences=True))
model.add(Dropout(0.1))

model.add(CuDNNLSTM(32))
model.add(Dropout(0.1))

model.add(Dense(32, activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(10, activation='softmax'))

opt = tf.keras.optimizers.Adam(lr=1e-3, decay=1e-5)

model.compile(
    loss='sparse_categorical_crossentropy',
    optimizer=opt,
    metrics=['accuracy'],
)

model.fit(x_train, y_train, epochs=3, callbacks=[tensorboard], validation_data=(x_test, y_test))
```
